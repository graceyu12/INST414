```python
import json
```

```python
import pandas as pd
```

```python
actor_genre = {}

with open("imdb_movies_2000to2022.prolific.json.", "r") as IMDB_file:
    for line in IMDB_file:
        movie = json.loads(line)
        genres = movie['genres']
        actors = movie['actors']

        for actor_id, actor_name in actors:
            if actor_name not in actor_genre:
                actor_genre[actor_name] = {}

            for genre in genres:
                if genre not in actor_genre[actor_name]:
                    actor_genre[actor_name][genre] = 0
                actor_genre[actor_name][genre] += 1
```

```python
index = actor_genre.keys()
rows = [actor_genre[k] for k in index]
df = pd.DataFrame(rows, index = index)
df = df.fillna(0)
df
```

```python
df_norm = df.divide(df.sum(axis=1), axis=0)
df_norm.head(10)
```

```python
import json
import random

import numpy as np
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

```python
g = nx.Graph()
```

```python
with open("imdb_movies_2000to2022.prolific.json", "r") as IMDB_file:
    for obj in IMDB_file:
        movie = json.loads(obj)
        if 'Drama' in movie.get('genres', []):
            actors = [actor[1] for actor in movie.get('actors', [])]
            for actor in actors:
                g.add_node(actor, name = actor)
            for i in range(len(actors)):
                for j in range(i + 1, len(actors)):
                    g.add_edge(actors[i], actors[j])
```

```python
centrality_degree = nx.degree_centrality(g)
top_k = 5
for u in sorted(centrality_degree, key = centrality_degree.get, reverse = True)[:top_k]:
    print(g.nodes[u]['name'], centrality_degree[u])
```

```python
target_actor_name = 'Eric Roberts'
target_actor = df.loc[target_actor_name]
distances = scipy.spatial.distance.cdist(df, [target_actor], metric = 'cosine').flatten()
query_distances = list(zip(df.index, distances))
for similar_actor_name, similar_genre_score in sorted(query_distances, key = lambda x:x[1], reverse=False)[:11]:
    print(similar_actor_name, similar_genre_score)
```
