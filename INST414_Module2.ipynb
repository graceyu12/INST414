```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("education_career_success.csv")
```

```python
df.head(5)
```

```python
new_df = df[["Field_of_Study","Starting_Salary"]]
new_df.head(5)
```

```python
new_df.isnull().values.any()
```

```python
fields = new_df["Field_of_Study"]
salaries = new_df["Starting_Salary"]
plt.bar(fields,salaries)
plt.xticks(rotation = 45)
plt.show()
```

```python
gpa_df = df[["Field_of_Study","University_GPA","Starting_Salary"]]
gpa_df['University_GPA'] = gpa_df['University_GPA'].round(0)
grouped_df = gpa_df.groupby(["Field_of_Study","University_GPA"])["Starting_Salary"].mean()
grouped_df.head(10)
```

```python
import seaborn as sns
sns.barplot(x = "Field_of_Study", y = "Starting_Salary", hue = "University_GPA", data = gpa_df)
plt.xticks(rotation = 45)
plt.show()
```
